```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('/content/compiled_results.csv')

# Filter the data to only keep rows where the "step" column is divisible by 8
df = df.dropna(subset=['validation_loss'])


# Calculate the mean of the "training loss" and "validation loss" columns for each 8-step interval
training_loss_mean = df.groupby('step')['training_loss'].mean()
validation_loss = df.groupby('step')['validation_loss'].mean()

# Plot the means using a line chart with "step" on the x-axis and "loss" on the y-axis
plt.figure(figsize=(12, 8))
plt.plot(training_loss_mean.index, training_loss_mean, label='Training Loss')
plt.plot(validation_loss.index, validation_loss, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#Load the CSV file into a pandas DataFrame
df = pd.read_csv('/content/compiled_results.csv')

#Filter the data to only keep rows where the "step" column is divisible by 8 and drop NaN validation loss
df = df.dropna(subset=['validation_loss'])

#Calculate the mean of the "training loss" and "validation loss" columns for each 8-step interval
training_loss_mean = df.groupby('step')['training_sequence_accuracy'].mean()
validation_loss = df.groupby('step')['validation_token_accuracy'].mean()

#Calculate the epoch by dividing the step by the number of steps per epoch (500)
epochs = df['step'] / 500

#Plot the means using a line chart with "epoch" on the x-axis and "loss" on the y-axis
plt.figure(figsize=(12, 8))
# plt.plot(epochs, training_loss_mean, label='training_sequence_accuracy')
plt.plot(epochs, validation_loss, label='Validation sequence accuracy')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
